```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
house = pd.read_csv("house.csv")
```

```python
house.head(10)
```

```python
house_sample = house.head(2000)
```

```python
print(house_sample)
```

```python
house_sample.drop("date",axis=1,inplace=True)
```

```python
house_sample.info()
```

```python
house_sample.corr()
```

# Line plot

```python
plt.figure()
sns.lineplot(house_sample['bedrooms'],house_sample['price'])
plt.show()
```

```python
plt.figure()
sns.scatterplot(data =house_sample,x = "bedrooms",y = "price")
plt.show()
```

## Histogram

```python
plt.figure()
sns.distplot(house_sample["bedrooms"])
plt.show()
```

## Bar Plot

```python
plt.figure()
plt.bar(house_sample["bedrooms"],house_sample["price"])
plt.show()
```

## PIE Chart

```python
plt.figure(figsize=(8,8))
house_sample['floors'].value_counts().plot(kind = 'pie',autopct = '%1.1f%%')
plt.show()
```
