# Visual_Assignment_1

### Try the following plots using seaborn.Lineplot, Scatter, Countplot, Histogram (Distplot) and subplots.

### lets use cars data

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Assignment_cars = pd.read_csv("cars.csv")
print(Assignment_cars)
```

# line plot

```python
plt.figure()
sns.lineplot(Assignment_cars["width"],Assignment_cars["height"])
plt.show()
```

# Scatterplot

```python
plt.figure()
sns.scatterplot(data = Assignment_cars,x="engine-size",y="price")
plt.show()
```

# Count plot

```python
plt.figure()
sns.countplot(data = Assignment_cars,x = "body-style")
plt.show()
```

```python
plt.figure()
sns.countplot(data = Assignment_cars, y = "body-style")
plt.show()
```

# Histogram

```python
plt.figure()
sns.distplot(Assignment_cars["width"])
plt.show()
```

# subplots

```python
fig,axes = plt.subplots(2,2,figsize=(12,10))

sns.countplot(data = Assignment_cars , x = "body-style", hue="fuel-type", ax = axes[0,0])

sns.scatterplot(data = Assignment_cars , x = "engine-size", y = "price", ax = axes[0,1])

sns.boxplot(data = Assignment_cars , x = "price", y = "body-style" , ax = axes[1,0])

sns.heatmap(Assignment_cars.corr() , annot = True , ax = axes[1,1])

plt.show()
```
