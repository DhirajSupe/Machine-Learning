```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Automobile_data.csv")
```

```python
data.head()
```

```python
df = data.drop(["aspiration","num-of-doors","wheel-base","length","curb-weight","num-of-cylinders","fuel-system","bore","stroke","compression-ratio","peak-rpm"], axis = 1, inplace = True)
```

```python
data
```

```python
# Replacing "?" sign from column "normalized-losses" with np.nan
data["normalized-losses"].replace("?",np.nan, inplace = True)

# coverting column "normalized losses" to float type
data["normalized-losses"] = data["normalized-losses"].astype(float)

# Taking mean of column "normalized-losses" 
losses_mean = data["normalized-losses"].mean()

# Replacing all np.nan with mean just calculated
data["normalized-losses"].fillna(losses_mean,inplace = True)
```

```python
# Doing same operations as above on column "horsepower"

data["horsepower"].replace("?",np.nan, inplace = True)

data["horsepower"] = data["horsepower"].astype(float)

horsepower_mean = data["horsepower"].mean()

data["horsepower"].fillna(horsepower_mean,inplace = True)
```

```python
data.dtypes
```

```python
data["price"].value_counts()
```

```python
# Doing same operations as above on column "price"

data["price"].replace("?",np.nan, inplace = True)

data["price"] = data["price"].astype(float)

price_mean = data["price"].mean()

data["price"].fillna(price_mean,inplace = True)
```

```python
data.dtypes
```

```python
data_num = data.select_dtypes(["int64","float64","int32"])
data_cat = data.select_dtypes(["object"])
```

```python
data_num.head()
```

```python
data_cat.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for col in data_cat:
    le = LabelEncoder()
    data_cat[col] = le.fit_transform(data_cat[col])
```

```python
data_cat.head()
```

```python
data_new = pd.concat([data_cat,data_num],axis = 1)
```

```python
data_new.head()
```

## Baseline Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = data_new.drop("price",axis = 1)
y = data_new["price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

```python
lin = LinearRegression()

lin.fit(X_train,y_train)
```

```python
lin.score(X_test, y_test)
```

```python
lin.coef_
```

## Lasso (L1) & Ridge (L2) Rgularization

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
for i in range(50):
    l2 = Ridge(alpha = i)
    l2.fit(X_train, y_train)
    print(i,":",l2.score(X_test,y_test))
```

```python
for i in range(200,500):
    l1 = Lasso(alpha = i)
    l1.fit(X_train, y_train)
    print(i,":",l1.score(X_test,y_test))
```

```python
l2 = Ridge(alpha = 11)
l2.fit(X_train, y_train)
print(l2.score(X_test, y_test))
```

```python
l2.coef_
```

```python
l1 = Lasso(alpha = 212)
l1.fit(X_train, y_train)
print(l1.score(X_test, y_test))
```

```python
l1.coef_
```

## K-fold Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
l1_cross = cross_val_score(l1,X,y,cv = 4)
print(l1_cross)
```
